```python
from sklearn.feature_extraction.text import TfidfVectorizer

def cluster_vectorizer(corpus):
    vect = TfidfVectorizer(sublinear_tf = True, min_df = 0.005)
    features = vect.fit_transform(corpus).toarray()
    return features, vect
```
